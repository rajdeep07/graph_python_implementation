```python
############################################################
# L0909.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2020
###########################################################
############################################################
# All imports
###########################################################
from typing import List
from time import process_time 
import collections
import queue
############################################################
#Read this for deque
#https://www.tutorialspoint.com/python/python_deque.htm
############################################################

class Solution:
    ## YOU CANNOT CHANGE THIS INTERFACE
    def snakesAndLadders(self, board: List[List[int]]) -> int:
        num_moves = [-1]
        work = [0]
        p = L0909(board,num_moves,work,False)
        return num_moves[0]
    
class QueueEntry:
        def __init__(self, v = 0, dist = 0):
            self.v = v
            self.dist = 0
    
class L0909:    
    def __init__(self, board: List[List[int]], num_moves:'List of size 1',work:'List of size 1',show:'boolean'):
        self._b = board 
        self._num_moves = num_moves
        self._work = work
        self._show = show
        num_row = len(self._b)
        num_col = len(self._b[0])
        assert(num_row == num_col)
        self._N = num_row
        num_moves = [-1]
        self._MAX = self._N * self._N
        self._alg()

    def _set_num_moves(self, m:'int'):
        self._num_moves[0] = m 

    def _increment_work(self):
        self._work[0] = self._work[0] + 1
        
    def _build_node_array(self):
        for i in range(0, self._MAX + 1):
            d = Data(i)
            self._node_array.append(d)
        
   
    def _alg(self):
        #MUST WRITE 2 routines
        #WILL NOT BE GRADED IF YOU CANNOT PRINT ALL PATHS
        if (self._show):
            self._print_matrix_with_coordinates()
        self._bfs()
        if (self._show):
            print("The paths from 1 to",self._MAX, "are as follows")
            self._print_all_shortest_path()
            print("WORK DONE =",self._work[0])
            
    def _build_data_structure(self):
        self._SL = {} # Dictionary of Snake and Ladder. Key is Start, Value is end.
        self._node_array = [] # list of all nodes
        self._queue = collections.deque() # queue
        
        self._build_SL_dictionary()
        self._build_node_array()
        
    def _print_matrix_with_coordinates(self):
        for r in range(0, self._N):
            for c in range(0, self._N):
                n = self._get_n(r,c)
                print("[",r,c,n,"]", sep=' ',end=' ')
                val = self._b[r][c]
                print(val, sep = ' ', end = ' ')
            print()
            
    def _get_rc(self, n:'int')->'List of row, col':
        a = [0, 0]
        assert(n > 0 and n <= self._MAX)
        r = (n-1) // self._N # quotient
        col = (n-1) % self._N # reminder
        row = self._N -1 -r
        if (r % 2):
            # ODD
            col = (self._N - 1) - col
        
        return [row, col]
    
    def _get_n(self, r:'int', c:'int')->'int':
        assert(r >= 0 and r < self._N)
        assert(c >= 0 and c < self._N)
        n = 0
        if (self._N % 2):
            #ODD
            if (r % 2):
                n = self._MAX - ((r * self._N) + c)
            else:
                n = self._MAX - ((r * self._N) + (self._N - 1 - c))
        else:
            ## EVEN
            if (r % 2):
                n = self._MAX - ((r * self._N) + (self._N - 1 - c))
            else:
                n = self._MAX - ((r * self._N) + c)
        
        assert(n > 0 and n <= self._MAX)
        return n
    
    def _is_ladder(self, n:'int')->'int':
        if n in self._SL.keys():
            k = self._SL[n]
            if (n< k):
                return True
        return False
    
    def _is_snake(self, n:'int')->'int':
        if n in self._SL.keys():
            k = self._SL[n]
            if (n > k):
                return True
        return False
    
    def _add_to_queue(self, d:'Data'):
        self._queue.append(d)
        
    def _is_queue_empty(self):
        if (len(self._queue) == 0):
            return True
        return False
    
    def _remove_from_queue(self):
        n = self._queue.pop()
        return n
    
    def _print_coordinates(self):
        j = 0
        for i in range(1, self._MAX + 1):
            print(i, self._get_rc(i))
            j = j + 1
            if (j == self._N):
                j = 0
                print("-----------")
                
    def _build_SL_dictionary(self):
        if (False):
            self._print_coordinates()
        for i in range(1, self._MAX + 1):
            [r,c] = self._get_rc(i)
            val = self._b[r][c]
            if (val != -1):
                self._SL[i] = val
            
        self._print_snake_and_ladder()
        
    def _print_snake_and_ladder(self):
        if (self._show):
            print("Snake and Ladder in the board are as follows.")
            for key, value in self._SL.items():
                if (key < value):
                    print("LADDER", key, "->", value)
                else:
                    print("SNAKE", key, "->", value)
    
    def _get_snake_or_ladder_or_nothing(self, n:'int') -> 'int':
        if n in self._SL.keys():
            return self._SL[n]
        return n
        
    def _print_all_shortest_path(self):
        for i in range(2, self._MAX + 1):
            if (self._is_snake(i) or self._is_ladder(i) ):
                print("1->",i,"No Path",sep="")
            else:
                self._print_A_shortest_path(i)
                
    def _print_A_shortest_path_r(self, d:'int'):
        d = self._node_array[d]._from
        t = self._node_array[d]._through
        if (d != 1):
            self._print_A_shortest_path_r(d)
            if (d != t):
                print("(", t, ")",sep = "",end="")
            print(d, "->", sep='',end='')
    
    def _print_A_shortest_path(self, tend:'int'):
        shortest_path = self._node_array[tend]._level
        if (shortest_path != -1):
            print("1->",sep='',end='')
            self._print_A_shortest_path_r(tend)
            print(tend, "Length =", shortest_path)
        
    def _bfs(self):
        self._build_data_structure()
        start = 1
        ns = self._node_array[start]
        ns._level = 0 # Not visited
        self._add_to_queue(ns)
        
        while (self._is_queue_empty() == False):
            n = self._remove_from_queue()
            
            for i in range(1,7):
                z = n._num + i
                if (z <= self._MAX):
                    nz = self._node_array[z]
                    nznum = nz._num
                    nznum2 = self._get_snake_or_ladder_or_nothing(nznum)
                    nzz = self._node_array[nznum2]
                    if (nzz._level != -1):
                        continue
                    self._increment_work()
                    nzz._from = n._num
                    nzz._level = 1 + n._level
                    nzz._through = nznum
                    self._add_to_queue(nzz)
        self._set_num_moves(self._node_array[self._MAX]._level)

class Data:
    def __init__(self, num):
        self._num = num
        self._level = -1 # Not visited
        self._from = -1
        self._through = -1
```
